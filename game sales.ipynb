```python
df=pd.read_csv("./gameSales.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting a histogram of sales
plt.figure(figsize=(10, 6))
sns.histplot(df['total_sales'], bins=20, kde=True)
plt.title('Distribution of Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Frequency')
plt.show()

# Creating a scatter plot of critic score vs. total sales
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='critic_score', y='total_sales')
plt.title('Critic Score vs. Total Sales')
plt.xlabel('Critic Score')
plt.ylabel('Total Sales')
plt.show()
```

```python

publisher_sales = df.groupby('publisher')['total_sales'].sum().sort_values(ascending=False)
print(publisher_sales)


platform_sales = df.groupby('title')['total_sales'].mean().sort_values(ascending=False)
print(platform_sales)
```

```python
# Box plot of sales by genre
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x='genre', y='total_sales')
plt.title('Total Sales by Genre')
plt.xlabel('Genre')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()

# Time series analysis of sales over years
df['release_date'] = pd.to_datetime(df['release_date'])
df['year'] = df['release_date'].dt.year
yearly_sales = df.groupby('year')['total_sales'].sum()

plt.figure(figsize=(12, 6))
yearly_sales.plot(marker='o')
plt.title('Total Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
genre_yearly_sales = df.groupby(['genre', 'year'])['total_sales']
genre_yearly_sales
```

```python
# Grouping data by genre and year
genre_yearly_sales = df.groupby(['genre', 'year'])['total_sales'].sum().reset_index()

# Plotting genre-wise sales over time
plt.figure(figsize=(12, 8))
sns.lineplot(data=genre_yearly_sales, x='year', y='total_sales', hue='genre', marker='o')
plt.title('Genre-wise Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.legend(title='Genre', loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```

```python
# Creating a bar plot to compare sales across regions
region_sales = df[['na_sales', 'jp_sales', 'pal_sales']].sum()
region_sales.plot(kind='bar', figsize=(10, 6))
plt.title('Total Sales Comparison Across Regions')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.show()
```

```python
# Extracting month from release date
df['month'] = df['release_date'].dt.month_name()

# Plotting the number of releases per month
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='month', order=df['month'].value_counts().index)
plt.title('Number of Game Releases by Month')
plt.xlabel('Month')
plt.ylabel('Number of Releases')
plt.xticks(rotation=45)
plt.show()
```

```python
# Identify non-numeric columns
non_numeric_columns = df.select_dtypes(exclude=['number']).columns
print(non_numeric_columns)
```

```python
# Drop non-numeric columns
numeric_df = df.drop(non_numeric_columns, axis=1)

# Calculate the correlation matrix
correlation_matrix = numeric_df.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
